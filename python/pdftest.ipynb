```python
# 폴더 내 jpg파일 전부 변환
import os
import pytesseract
from PIL import Image
import re

# Tesseract 경로 설정
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'

# 이미지가 있는 폴더 경로
folder_path = 'C:/Users/infodea nb510/Documents/유현준/python/image45_66/'

# 파일명에서 숫자를 추출하는 함수
def extract_number(filename):
    # 파일명에서 숫자 부분만 추출하여 반환
    numbers = re.findall(r'\d+', filename)
    return int(numbers[0]) if numbers else 0  # 숫자가 없으면 0 반환

# 폴더 내 모든 파일을 숫자 기준으로 정렬
files = sorted(os.listdir(folder_path), key=extract_number)

# 파일 순서대로 OCR 처리
for filename in files:
    if filename.endswith('.jpg'):
        image_path = os.path.join(folder_path, filename)

        # 이미지 열기
        img = Image.open(image_path)

        # OCR 처리
        text = pytesseract.image_to_string(img, lang='kor+eng', config='-c preserve_interword_spaces=1 --psm 4')

        # 출력
        print(f"파일명: {filename}")
        print(text)
        print("/"*50)  # 구분선
```
